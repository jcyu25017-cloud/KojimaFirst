```python
x = 10
y = 20
print('xはyより大きい')
if (x > y):
    print('True')
else:
    print('False')
print('xはy以上')
if (x >= y):
    print('True')
else:
    print('False')
print('xはyと等しい')
if (x == y):
    print('True')
else:
    print('False')
print('xはy以下')
if (x <= y):
    print('True')
else:
    print('False')
print('xはyより小さい')
if (x < y):
    print('True')
else:
    print('False')
```
