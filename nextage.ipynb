```python
st = {1,2,3}
print(st)
st.add(1)
print(st)

st.remove(1)
st.discard(1) #存在しない要素を削除してもエラーにならない
print(st)
```

```python
math_members ={1, 2, 4, 5, 7}
eng_members ={2, 3, 5, 8}
targets = []
for i in range(30):
     if (i % 3 == 0):
        continue
     targets.append(i)
     if (i >= 20):
        break
print('対象者:', targets)
print('数学または英語:', math_members | eng_members)
print('両方', math_members & eng_members)
```
