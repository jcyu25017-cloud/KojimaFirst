```python
prices = {
    'apple': 100,
    'banana': 80,
    'orange': 120
}

order_text = "apple,banana,orange"

lst = order_text.split(',')
#print(lst)
sum = 0
for fruits, price in prices.items():
    if fruits in lst:
        sum = sum + price
#print(f"合計: {sum}円")
name = r"/".join(lst)

try:
    print(f"注文内容: {name}")
    print(f"合計金額: {sum}円")
except KeyError:
    print("エラーが発生しました:")#このままだとリストに同一商品があった場合金額が合わない
```

```python
# 1. 商品名と価格を辞書に用意
prices = {
    "apple": 100,
    "banana": 80,
    "orange": 120
}

order_text = "apple,banana,orange"

try:
    # 2. 注文データを「,」で分割してリストにする
    order_items = order_text.split(',')

    # 3. 注文された商品の価格を辞書から取り出し、リストにする (ヒント3)
    # 辞書にないキーがあると、ここで KeyError が発生します (ヒント6)
    price_list = [prices[item] for item in order_items]

    # 4. すべての価格を合計する (ヒント4)
    total_price = 0
    for price in price_list:
        total_price += price

    # 5. 注文内容を apple / banana / orange の形式で表示 (ヒント5)
    print(f"注文内容: {' / '.join(order_items)}")

    # 6. 合計金額を表示
    print(f"合計金額: {total_price}円")

except KeyError as e:
    # 7. 存在しない商品名が入っていた場合のエラー処理
    print(f"エラー: 商品 {e} はメニューに存在しません。")
```

```python
prices = {
    'apple': 100,
    'banana': 80,
    'orange': 120
}

order_text = "apple,apple,orange"

lst = order_text.split(',')

try:
    price_list = [prices[item] for item in lst]
    total_price = 0
    for price in price_list:
        total_price += price
    print(f"注文内容: {' / '.join(lst)}")
    print(f"合計金額: {total_price}円")

except KeyError:
    print(f"エラー: 商品 はメニューに存在しません。")

#sumは使うべきでない
```

prices は「本のタイトル」
最初に prices = { ... } と書いたとき、あなたは prices という名前の「価格表（辞書）」を作りました。
これは、すべてのデータが詰まった「本全体」を指しています。

2. prices[item] は「中身の検索」
prices[item] の [ ]（角カッコ）は、Pythonでは「検索する」という意味の記号です。

もし、ループの中で item が "apple" だった場合：

prices[item] は prices["apple"] と同じ意味になります。

すると、Pythonは prices という名前の本の中から "apple" のページを探し、100 という数字を引っ張り出してきます
